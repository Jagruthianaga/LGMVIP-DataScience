```python
import cv2
import matplotlib.pyplot as plt
```

**Loading the Image...**

```python
img=cv2.imread("Mickey.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.axis("off")
plt.title("Mickey Image")
plt.show()
```

**Converting the Original Image to GrayScale Image:**

```python
gray_image=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

**Inverting the Gray Scale Image**

```python
inverted_gray_image=255-gray_image
plt.figure(figsize=(10,10))
plt.imshow(inverted_gray_image,cmap="gray")
plt.axis("off")
plt.title("Inverted GrayScale Image")
plt.show()
```

**Converting Inverting into Blurred Image** 

```python
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21), 0)
plt.figure(figsize=(10,10))
plt.imshow(blurred_img,cmap="gray")
plt.axis("off")
plt.title("Blurred Image")
plt.show()
```

**Inverting the Blurred one**

```python
inverted_blurred_image = 255-blurred_img
plt.figure(figsize=(10,10))
plt.imshow(inverted_blurred_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Blurred Image")
plt.show()
```

**Image to pencil Sketch of Mickey Mouse**

```python
pencil_sketch_image = cv2.divide(gray_image,inverted_blurred_image,scale=256.0)
plt.figure(figsize=(10,10))
plt.imshow(pencil_sketch_image,cmap="gray")
plt.axis("off")
plt.title("Pencil Sketch Image")
plt.show()
```

**Steps of Converting an Image to Pencil Sketch**


```python
plt.figure(figsize=(20,20))
plt.subplot(1,7,1)
plt.imshow(img)
plt.axis("off")
plt.title("Mickey Mouse")
plt.subplot(1,7,2)
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,7,3)
plt.imshow(inverted_gray_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,7,4)
plt.imshow(blurred_img,cmap="gray")
plt.axis("off")
plt.title("Blurred Image")
plt.subplot(1,7,5)
plt.imshow(inverted_blurred_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Blurred Image")
plt.subplot(1,7,6)
plt.imshow(pencil_sketch_image,cmap="gray")
plt.axis("off")
plt.title("Image to Pencil Sketch Image")
plt.show()
```
